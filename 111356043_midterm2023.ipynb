```
# def softmax(x):
#   return np.exp(x)/np.sum(np.exp(x))
# def cross_entropy1(x,y):
#   x_softmax = [softmax(x[i]) for i in range(len(x))]
#   x_log = [np.log(x_softmax[i][y[i]]) for i in range(len(y))]
#   loss = -np.sum(x_log)/len(y)
#   return loss
def cross_entropy(y_true, y_pred):
  # y_true = np.array([0, 1, 1, 0, 1])
  # y_pred = np.array([0.1, 0.9, 0.8, 0.2, 0.7])
  cross = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
  return cross
```

這時候，用混肴矩陣就可以解決這樣資料不均的問題，因為混淆矩陣分為四象限，包含TP, FP, TN, FN，所以他可以真正測出是不是矇對的，另外也可以用這些參數計算recall, F1-score, ROC-AUC、面積等等。
